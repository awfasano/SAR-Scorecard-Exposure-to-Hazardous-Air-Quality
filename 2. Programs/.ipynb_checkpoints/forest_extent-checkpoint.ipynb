```python
import os
import glob
import rasterio
from rasterio.merge import merge

# Find all raster files in the specified folder
raster_files = sorted(glob.glob(os.path.join(raster_folder, '*.tif')))

# Check if any raster files are found
if not raster_files:
    raise FileNotFoundError(f"No raster files found in {raster_folder}")

print(f"Found {len(raster_files)} raster files. Creating mosaic...")

# Open raster files safely
src_files_to_mosaic = []
for fp in raster_files:
    try:
        src = rasterio.open(fp)
        src_files_to_mosaic.append(src)
    except Exception as e:
        print(f"Skipping {fp} due to error: {e}")

# Ensure at least one valid raster is available
if not src_files_to_mosaic:
    raise RuntimeError("No valid raster files could be opened.")

# Merge the raster files
mosaic, out_trans = merge(src_files_to_mosaic)

# Get metadata of the first raster file (assuming all rasters have the same metadata)
out_meta = src_files_to_mosaic[0].meta.copy()

# Update metadata with new dimensions, transform (affine), and CRS
out_meta.update({
    "driver": "GTiff",
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "transform": out_trans,
    "crs": src_files_to_mosaic[0].crs
})

# Ensure correct band writing (multi-band handling)
with rasterio.open(output_raster, "w", **out_meta) as dest:
    dest.write(mosaic)

print(f"Mosaic raster saved at: {output_raster}")

# Close raster files
for src in src_files_to_mosaic:
    src.close()
```

```python
import os
import rasterio
import numpy as np

# Paths
input_raster = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\lclc_2020.tif"  # Created mosaic
output_raster = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\forest_mask_2020.tif"  # Mask output

# Define NoData value
nodata_value = 255  # NoData value to be assigned to non-forest areas

# Open the mosaic raster
with rasterio.open(input_raster) as src:
    # Read the data
    landcover = src.read(1)  # Assuming single-band raster

    # Create forest mask: Keep forest values, assign NoData to non-forest areas
    forest_mask = np.where(
        ((landcover >= 25) & (landcover <= 48)) | ((landcover >= 125) & (landcover <= 148)), 
        1,  # Forest (value 1)
        nodata_value  # NoData (255)
    ).astype(np.uint8)  # Convert to 8-bit integer

    # Copy metadata and update for mask
    mask_meta = src.meta.copy()
    mask_meta.update({
        "dtype": "uint8",  # Single-band 8-bit raster
        "count": 1,  # One band
        "nodata": nodata_value  # Set NoData value
    })

    # Write the mask raster
    with rasterio.open(output_raster, "w", **mask_meta) as dst:
        dst.write(forest_mask, 1)

print(f"Forest mask raster saved at: {output_raster}")
```

```python
import os
import rasterio
import geopandas as gpd
import numpy as np
import pandas as pd
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio.mask import mask

# Paths to input data

forest_mask_raster = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\forest_mask_2020.tif"
reprojected_forest_mask = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\forest_mask_2020_reprojected.tif"

# Define target CRS (EPSG:7780 - South Asia Albers Equal Area)

target_crs = "EPSG:7780"
nodata_value = 255  # NoData value

# Function to reproject raster

def reproject_raster(src_path, dst_path, target_crs):
    with rasterio.open(src_path) as src:
        transform, width, height = calculate_default_transform(
            src.crs, target_crs, src.width, src.height, *src.bounds)
        kwargs = src.meta.copy()
        kwargs.update({
            'crs': target_crs,
            'transform': transform,
            'width': width,
            'height': height
        })
        with rasterio.open(dst_path, 'w', **kwargs) as dst:
            for i in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=target_crs,
                    resampling=Resampling.nearest)
        print(f"Reprojected forest mask saved: {dst_path}")

# Function to reproject shapefile

def reproject_shapefile(src_path, target_crs):
    gdf = gpd.read_file(src_path)
    gdf = gdf.to_crs(target_crs)
    return gdf  # Return the reprojected GeoDataFrame

# Function to calculate area in **millions of hectares**
def calculate_area(geom):
    return geom.area / 10**8  # Convert from m² to million hectares


# **Reproject the forest mask raster**
reproject_raster(forest_mask_raster, reprojected_forest_mask, target_crs)
```

```python
import os
import rasterio
import geopandas as gpd
import numpy as np
import pandas as pd
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio.mask import mask

# Paths to input data
sar_shp0 = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Shapefile\WB_GAD\WB_GAD_ADM0_SAR_Clean.shp"
reprojected_forest_mask = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\forest_mask_2020_reprojected.tif"
output_csv_full = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\3. Output\Forest\forest_area_by_admin0_full.csv"
output_csv_agg = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\3. Output\Forest\forest_area_by_admin0_aggregated.csv"


# Load and reproject the shapefile (sar_shp0)
shp0_df = reproject_shapefile(sar_shp0, target_crs)

# Check if "globalid" exists
if "globalid" not in shp0_df.columns:
    raise KeyError("The shapefile does not contain a 'globalid' field.")


results = []

# Open and process the reprojected forest mask
with rasterio.open(reprojected_forest_mask) as src:
    raster_transform = src.transform
    raster_crs = src.crs

    # Iterate over each zone (polygon) in the shapefile
    for idx, row in shp0_df.iterrows():
        try:
            # Mask the raster with the current polygon
            out_image, out_transform = mask(src, [row.geometry], crop=True)
            out_image = out_image[0]  # Single band

            # Compute total area of the zone (millions of hectares)
            total_area = calculate_area(row.geometry)

            # Ensure NoData values (255) are ignored
            valid_pixels = out_image[out_image != nodata_value]

            # Compute forest area (sum of valid forest pixels)
            forest_area = (valid_pixels == 1).sum() * out_transform[0]**2 / 10**8

            # Append results 
            results.append({
                'globalid': row['globalid'],  
                'total_area': total_area,
                'forest_area': forest_area,
                'geo_level': 0  # Set geo_level = 0
            })
        except Exception as e:
            print(f"Error processing polygon {idx}: {e}")

# Convert results to DataFrame
df = pd.DataFrame(results)

# **Merge shp0_df with computed areas (df) on globalid**
merged_df = shp0_df.merge(df, on='globalid', how='left')

# **Remove geometry before saving**
merged_df = merged_df.drop(columns=['geometry'], errors='ignore')

# Save full dataset before aggregation
merged_df.to_csv(output_csv_full, index=False)
print(f"Full dataset saved to {output_csv_full}")

# **Aggregate by L0_CODE** (since multiple polygons may belong to the same country)
agg_df = merged_df.groupby(['L0_CODE']).agg({
    'L0_NAME' :  'first' ,
    'total_area': 'sum',
    'forest_area': 'sum',
    'geo_level': 'first',  # Keep geo_level = 0
    'wb_status': 'first',
    'sovereign': 'first',
    'Disputed' : 'first'
}).reset_index()

# Save Aggregated CSV
agg_df.to_csv(output_csv_agg, index=False)
print(f"Aggregated dataset saved to {output_csv_agg}")

```

```python
import os
import rasterio
import geopandas as gpd
import numpy as np
import pandas as pd
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio.mask import mask

# Paths to input data
sar_shp0 = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Shapefile\WB_GAD\WB_GAD_ADM0_SAR_Clean.shp"
reprojected_forest_mask = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\forest_mask_2010_reprojected.tif"
output_csv_full = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\3. Output\Forest\2010\forest_area_by_admin0_full.csv"
output_csv_agg = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\3. Output\Forest\2010\forest_area_by_admin0_aggregated.csv"


# Load and reproject the shapefile (sar_shp0)
shp0_df = reproject_shapefile(sar_shp0, target_crs)

# Check if "globalid" exists
if "globalid" not in shp0_df.columns:
    raise KeyError("The shapefile does not contain a 'globalid' field.")


results = []

# Open and process the reprojected forest mask
with rasterio.open(reprojected_forest_mask) as src:
    raster_transform = src.transform
    raster_crs = src.crs

    # Iterate over each zone (polygon) in the shapefile
    for idx, row in shp0_df.iterrows():
        try:
            # Mask the raster with the current polygon
            out_image, out_transform = mask(src, [row.geometry], crop=True)
            out_image = out_image[0]  # Single band

            # Compute total area of the zone (millions of hectares)
            total_area = calculate_area(row.geometry)

            # Ensure NoData values (255) are ignored
            valid_pixels = out_image[out_image != nodata_value]

            # Compute forest area (sum of valid forest pixels)
            forest_area = (valid_pixels == 1).sum() * out_transform[0]**2 / 10**8

            # Append results 
            results.append({
                'globalid': row['globalid'],  
                'total_area': total_area,
                'forest_area': forest_area,
                'geo_level': 0  # Set geo_level = 0
            })
        except Exception as e:
            print(f"Error processing polygon {idx}: {e}")

# Convert results to DataFrame
df = pd.DataFrame(results)

# **Merge shp0_df with computed areas (df) on globalid**
merged_df = shp0_df.merge(df, on='globalid', how='left')

# **Remove geometry before saving**
merged_df = merged_df.drop(columns=['geometry'], errors='ignore')

# Save full dataset before aggregation
merged_df.to_csv(output_csv_full, index=False)
print(f"Full dataset saved to {output_csv_full}")

# **Aggregate by L0_CODE** (since multiple polygons may belong to the same country)
agg_df = merged_df.groupby(['L0_CODE']).agg({
    'L0_NAME' :  'first' ,
    'total_area': 'sum',
    'forest_area': 'sum',
    'geo_level': 'first',  # Keep geo_level = 0
    'wb_status': 'first',
    'sovereign': 'first',
    'Disputed' : 'first'
}).reset_index()

# Save Aggregated CSV
agg_df.to_csv(output_csv_agg, index=False)
print(f"Aggregated dataset saved to {output_csv_agg}")

```

```python
import os
import glob
import rasterio
from rasterio.merge import merge

# Find all raster files in the specified folder
raster_files = sorted(glob.glob(os.path.join(raster_folder, '*.tif')))

# Check if any raster files are found
if not raster_files:
    raise FileNotFoundError(f"No raster files found in {raster_folder}")

print(f"Found {len(raster_files)} raster files. Creating mosaic...")

# Open raster files safely
src_files_to_mosaic = []
for fp in raster_files:
    try:
        src = rasterio.open(fp)
        src_files_to_mosaic.append(src)
    except Exception as e:
        print(f"Skipping {fp} due to error: {e}")

# Ensure at least one valid raster is available
if not src_files_to_mosaic:
    raise RuntimeError("No valid raster files could be opened.")

# Merge the raster files
mosaic, out_trans = merge(src_files_to_mosaic)

# Get metadata of the first raster file (assuming all rasters have the same metadata)
out_meta = src_files_to_mosaic[0].meta.copy()

# Update metadata with new dimensions, transform (affine), and CRS
out_meta.update({
    "driver": "GTiff",
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "transform": out_trans,
    "crs": src_files_to_mosaic[0].crs
})

# Ensure correct band writing (multi-band handling)
with rasterio.open(output_raster, "w", **out_meta) as dest:
    dest.write(mosaic)

print(f"Mosaic raster saved at: {output_raster}")

# Close raster files
for src in src_files_to_mosaic:
    src.close()
```

```python
import os
import rasterio
import numpy as np

# Paths
input_raster = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\lclc_2015.tif"  # Created mosaic
output_raster = r"C:\Users\vgald\OneDrive\Desktop\SAR_DATA\1. Data\Forest\forest_mask_2015.tif"  # Mask output

# Define NoData value
nodata_value = 255  # NoData value to be assigned to non-forest areas

# Open the mosaic raster
with rasterio.open(input_raster) as src:
    # Read the data
    landcover = src.read(1)  # Assuming single-band raster

    # Create forest mask: Keep forest values, assign NoData to non-forest areas
    forest_mask = np.where(
        ((landcover >= 25) & (landcover <= 48)) | ((landcover >= 125) & (landcover <= 148)), 
        1,  # Forest (value 1)
        nodata_value  # NoData (255)
    ).astype(np.uint8)  # Convert to 8-bit integer

    # Copy metadata and update for mask
    mask_meta = src.meta.copy()
    mask_meta.update({
        "dtype": "uint8",  # Single-band 8-bit raster
        "count": 1,  # One band
        "nodata": nodata_value  # Set NoData value
    })

    # Write the mask raster
    with rasterio.open(output_raster, "w", **mask_meta) as dst:
        dst.write(forest_mask, 1)

print(f"Forest mask raster saved at: {output_raster}")
```
